```python
import pandas as pd
```

```python
"""
Dataset: https://footystats.org/download-stats-csv
"""

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('data/england-premier-league-matches-2018-to-2019-stats.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
def determine_win(row):
  if row['home_team_goal_count'] > row['away_team_goal_count']:
    return 1
  elif row['home_team_goal_count'] < row['away_team_goal_count']:
    return -1
  else:
    return 0

df['win'] = df.apply(determine_win, axis=1)
```

```python
df.loc[:, ['win', 'home_team_goal_count', 'away_team_goal_count']].head()
```

```python
half_time_columns = ['home_team_goal_count_half_time', 'away_team_goal_count_half_time']
df[half_time_columns].head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import classification_report

models = [
  DecisionTreeClassifier,
  GradientBoostingClassifier,
  RandomForestClassifier,
  LogisticRegression,
  KNeighborsClassifier,
  SVC,
  GaussianNB,
  MLPClassifier,
  AdaBoostClassifier,
  LinearDiscriminantAnalysis,
  BaggingClassifier
]

def run_all_classifiers(features):
  for model in models:
    run_classifier(features, model)

def run_classifier(features, model):
  # Define the features and target variables
  target = 'win'

  # Split the data into training and testing sets
  X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

  # Create the classifier
  clf = model()

  # Train the model
  clf.fit(X_train, y_train)

  # Make predictions on the test set
  y_pred = clf.predict(X_test)

  # Compute other classification metrics
  report = classification_report(y_test, y_pred)

  # Print the results
  print(f'{model.__name__} results:')
  print(report)
```

```python
run_all_classifiers(['home_team_goal_count', 'away_team_goal_count'])
```

```python
run_all_classifiers(['home_team_goal_count_half_time', 'away_team_goal_count_half_time'])
```

```python
important_features = [
    "Pre-Match PPG (Home)", "Pre-Match PPG (Away)",
    "home_team_shots_on_target", "away_team_shots_on_target",
    "home_team_possession", "away_team_possession",
    "Home Team Pre-Match xG", "Away Team Pre-Match xG",
    "home_team_corner_count", "away_team_corner_count",
    "home_team_yellow_cards", "home_team_red_cards",
    "away_team_yellow_cards", "away_team_red_cards",
]

run_all_classifiers(important_features)
```
