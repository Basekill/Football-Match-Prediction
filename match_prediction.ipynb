```python
drop_columns = ['btts_percentage_pre_match', 'over_15_percentage_pre_match', 'over_25_percentage_pre_match', 'over_35_percentage_pre_match', 'over_45_percentage_pre_match', 'over_15_HT_FHG_percentage_pre_match', 'over_05_HT_FHG_percentage_pre_match', 'over_15_2HG_percentage_pre_match', 'over_05_2HG_percentage_pre_match', 'average_corners_per_match_pre_match', 'average_cards_per_match_pre_match', 'odds_ft_home_team_win', 'odds_ft_draw', 'odds_ft_away_team_win', 'odds_ft_over15', 'odds_ft_over25', 'odds_ft_over35', 'odds_ft_over45', 'odds_btts_yes', 'odds_btts_no', 'stadium_name']

unused_columns = [
  'timestamp',
  'date_GMT',
  'status',
  'attendance',
  'referee',
  'Game Week',
  # 'Pre-Match PPG (Home)',
  # 'Pre-Match PPG (Away)',
  'Home Team Pre-Match xG',
  'Away Team Pre-Match xG',
  'average_goals_per_match_pre_match',
]
drop_columns.extend(unused_columns)

df.drop(drop_columns, axis=1, inplace=True)
```

```python
df.columns
```

```python
def determine_win(row):
  if row['home_team_goal_count'] > row['away_team_goal_count']:
    return 1
  elif row['home_team_goal_count'] < row['away_team_goal_count']:
    return -1
  else:
    return 0

df['win'] = df.apply(determine_win, axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from joblib import dump, load

models = [
  DecisionTreeClassifier,
  GradientBoostingClassifier,
  RandomForestClassifier,
  LogisticRegression,
  KNeighborsClassifier,
  SVC,
  GaussianNB,
  MLPClassifier,
  AdaBoostClassifier,
  LinearDiscriminantAnalysis,
  BaggingClassifier
]

def run_all_classifiers(features, save=True):
  highest_f1_score = 0
  best_classifier = None

  for model in models:
    classifier, results = run_classifier(features, model)
    f1_score = results['macro avg']['f1-score']
    
    if f1_score > highest_f1_score:
      highest_f1_score = f1_score
      best_classifier = classifier
  
  if save:
    dump(best_classifier, 'models/best_classifier.joblib')

  return best_classifier, highest_f1_score

def run_classifier(features, model):
  # Define the features and target variables
  target = 'win'

  # Split the data into training and testing sets
  X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

  # Create the classifier
  clf = model()

  # Train the model
  clf.fit(X_train, y_train)

  # Make predictions on the test set
  y_pred = clf.predict(X_test)

  # Compute other classification metrics
  report = classification_report(y_test, y_pred, output_dict=True)

  # Print the results
  print(f'{model.__name__} results:')
  print(report)

  # Return the classification report
  return clf, report
```

```python
run_all_classifiers(['home_team_goal_count', 'away_team_goal_count'])
```

```python
run_all_classifiers(['home_team_goal_count_half_time', 'away_team_goal_count_half_time'])
```
